```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Ecommerce Customers.csv')
```

```python
df
```

```python
df.info()
```

```python
df.drop(['Email','Address','Avatar'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.corr()
```

```python
y=(df['Yearly Amount Spent'])
```

```python
y.head()
```

```python
df.drop('Yearly Amount Spent',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df,y,test_size=0.2)
```

```python
l_reg=LinearRegression()
```

```python
l_reg.fit(X_train,Y_train)
```

```python
print(l_reg.coef_)
print(l_reg.intercept_)
```

```python
y_pred=l_reg.predict(X_train)
```

```python
l_reg.score(X_train,Y_train)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=l_reg, X=X_train, y=Y_train, cv=10)
model_accuracy = accuracies.mean()
model_standard_deviation = accuracies.std()
```
